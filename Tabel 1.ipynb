```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Blood IL-6 Concentration
treatments = [
    "Kolesterol (PI)", "Meat + Max ds (PII)", "Meat + opt ds (PIII)", 
    "Meat +min ds (PIV)", "Meat only (PV)", "Rat food (PVI)"
]
weeks = ["Week -3", "Week -12", "Week -20"]
concentration_means = [
    [131.9, 140.5, 145.2],
    [110.0, 131.98, 134.8],
    [83.8, 130.6, 138.0],
    [94.0, 126.6, 139.6],
    [118.8, 129.4, 156.8],
    [151.5, 122.1, 152.3]
]

# Data for Blood IL-6 Expression
expression_means = [
    [74, 66, 156],
    [45, 48, 86],
    [58, 50, 57],
    [49, 50, 44],
    [57, 56, 71],
    [22, 28, 28]
]

# Enhanced Plot for Blood IL-6 Concentration with markers and data labels
plt.figure(figsize=(14, 6))
for i, (treatment, data) in enumerate(zip(treatments, concentration_means)):
    plt.plot(weeks, data, marker='o', label=treatment)
    for j, value in enumerate(data):
        plt.text(j, value + 2, f"{value:.1f}", ha='center', fontsize=9)

plt.title("Blood IL-6 Concentration")
plt.xlabel("Weeks")
plt.ylabel("Concentration (mean  SD)")
plt.legend(loc="upper left")
plt.grid(alpha=0.5)
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()

# Enhanced Plot for Blood IL-6 Expression with markers and data labels
plt.figure(figsize=(14, 6))
for i, (treatment, data) in enumerate(zip(treatments, expression_means)):
    plt.plot(weeks, data, marker='o', label=treatment)
    for j, value in enumerate(data):
        plt.text(j, value + 2, f"{value:.1f}", ha='center', fontsize=9)

plt.title("Blood IL-6 Expression")
plt.xlabel("Weeks")
plt.ylabel("Expression (mean  SD)")
plt.legend(loc="upper left")
plt.grid(alpha=0.5)
plt.tight_layout()
plt.show()

```
