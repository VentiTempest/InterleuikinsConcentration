```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define the data
data = {
    "Parameter": ["Berat Badan (g)", "F2-Isoprostan (pg/mL)", "IL-6 (pg/mL)", 
                  "Antioksidan Total (l/dL)", "Kadar GSH (mol/L)", 
                  "Jumlah Sel Busa", "Eksprei F2-isoprostan", "Ekspresi IL-6"],
    "Data Baseline": [121.40, 404.90, 128.90, 770.50, 1052.0, 0.0, 0.0, 0.0],
    "SD": [9.30, 9.70, 15.40, 81.50, 64.50, 0.0, 0.0, 0.0],
    "P": [0.134, 0.836, 0.986, 0.793, 0.672, None, None, None]
}

# Convert the data to a pandas DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x="Parameter", y="Data Baseline", data=df, palette="viridis", errorbar=None)

# Add error bars (standard deviation)
plt.errorbar(df["Parameter"], df["Data Baseline"], yerr=df["SD"], fmt='none', 
             color='black', capsize=5, label='SD')

# Annotating the P-values
for i, p_value in enumerate(df["P"]):
    if p_value is not None:  # Avoid annotating None values
        plt.text(i, df["Data Baseline"][i] + 10, f"P = {p_value:.3f}",
                 ha='center', va='bottom', fontsize=10)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Adding labels and title
plt.xlabel('Parameter')
plt.ylabel('Data Baseline')
plt.title('Data Baseline with SD and P-values')

# Show the plot
plt.tight_layout()
plt.show()

```

```python

import numpy as np

# Data for the two comparisons
parameters = ['IL-6 Baseline', 'WK 0 - WK III']
mean_IL6 = [128.9, 128.9]  # IL-6 Baseline Mean values
mean_WK = [128.9, 131.9]   # WK 0 - WK III Mean values
p_values = [0.986, 0.768]  # p-values for IL-6 and WK 0 - WK III comparisons

# Plotting
plt.figure(figsize=(8, 6))

# Plotting the lines for the two comparisons (with line connecting points for IL-6 Baseline and WK 0 - WK III)
plt.plot(parameters, mean_IL6, marker='o', linestyle='-', color='blue', label='IL-6 Baseline', linewidth=2)
plt.plot(parameters, mean_WK, marker='o', linestyle='-', color='green', label='WK 0 - WK III', linewidth=2)

# Annotating the Mean values for IL-6 Baseline and WK 0 - WK III
for i, mean in enumerate(mean_IL6):
    plt.text(i, mean + 0.5, f'{mean:.2f}', ha='center', va='bottom', fontsize=12)
for i, mean in enumerate(mean_WK):
    plt.text(i, mean + 0.5, f'{mean:.2f}', ha='center', va='bottom', fontsize=12)

# Annotating the p-values inside the plot area
plt.text(0, 128.9 + 1.5, f'P = {p_values[0]:.3f}', ha='center', va='bottom', fontsize=12, fontweight='bold', color='blue')
plt.text(1, 130.0, f'P = {p_values[1]:.3f}', ha='center', va='bottom', fontsize=12, fontweight='bold', color='green')

# Adding labels and title
plt.xlabel('Comparison')
plt.ylabel('Mean Value (pg/mL)')
plt.title('Comparison of IL-6 Baseline vs WK 0 - WK III')

# Displaying grid
plt.grid(True)

# Adding a legend to differentiate the two comparisons
plt.legend()

# Adjust y-axis limits to ensure p-values fit inside the graph area
plt.ylim(128, 132)

# Show the plot
plt.tight_layout()
plt.show()

```
